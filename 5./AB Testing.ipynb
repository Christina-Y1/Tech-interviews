### Задание
На экране покупки после онбординга приложения  конверсия в 3 месячную подписку равна 7,33%.  
Минимальный видимый эффект для этого показателя - 20%.   
Необходимо протестировать новый экран покупки, который должен повлиять на конверсию в 3-месячную подписку с помощью АБ теста. 

1. Рассчитайте необходимую выборку для проведения АБ теста чтобы данных было достаточно для подведения результатов. Объясните как считали.

2. По результатам АБ теста. В группе А - 2433 установок и 175 конверсий, в группе Б - 2479 установок, 210 конверсий. Принимаем вариант Б? Почему да/нет?"				

**Гипотезы**

Нужно определить две гипотезы, которые помогут понять, является ли наблюдаемая разница между версией A (изначальной) и версией B (новой, которую мы хотим проверить) случайностью или результатом изменений, которые мы произвели.

1. Нулевая. Изменения экрана покупки не дадут изменений, конверсия останется прежней (задача — опровергнуть гипотезу).  
2. Альтернативная. Изменения приведут к повышению конверсии.

### 1. Рассчитайте необходимую выборку для проведения АБ теста

**Группы**  
* Контрольная группа - группа, видящая версию A (экран покупки без изменений)  
* Эспериментальная группа - группа, видящая версию B (новый экран покупки)  

Обычно эти группы одинакового размера.

**Минимальный размер выборки** — чтобы результаты теста были статистически значимыми и им можно было доверять, необходимо определить количество пользователей, которые должны участвовать в тестировании. 

Для того, чтобы определить необходимый размер выборки - надо ориентироваться на минимальный видимый эффект и базовую конверсию.
Сделать это можно с помощью онлайн-калькуляторов, например: Driverback или Optimizely.  
**Optimizely** - показывает, что нам необходимо 4300 уникальных посетителей в каждую группу при уровне значимости 95%

**Driverback, Evanmiller** - Введя эти цифры в калькулятор мы видим, что для тестирования нам нужно минимум по 5094 уникальных посетителей в каждую группу

### 2. Принимаем вариант Б? Почему да/нет?

Если р-значение ниже заданного уровня значимости (альфа), тогда мы отвергаем нулевую гипотезу и выбираем альтернативную гипотезу.  
Уровень a - 5%

p = 0.096   
https://www.evanmiller.org/ab-testing/chi-squared.html#!175/2433;210/2479@95

Р-значение 0,096 означает, что есть 9% вероятность, что результат теста обусловлен случайностью.  
Как упоминалось ранее, мы установили альфа равным 0,05, прежде чем мы начали эксперимент, полученный результат не является статистически значимым, поскольку р-значение 0,096 выше, чем альфа.

Поэтому мы не отвергаем нулевую гипотезу - изменения экрана покупки не повлияют на конверсию в покупку трехмесячной подписки. Принимаем вариант А.
